### <font color=darkOrange>**1.5 Definición de hipótesis y supuestos**</font>

>Un supuesto muy grande es que nuestra compañía **ya estaría establecida**, es decir, no tendríamos mucha competencia y tendríamos un portafolio de propiedades en venta relativamente grande. Sin embargo, esto no significa administraríamos todas las propiedades que se encuentran en venta, por lo que también habrá que de cierta manera suponer y simular una **probabilidad de que la venta sea administrada por nosotros o no.** Es decir, tenemos cierta participación en el mercado pero no todas las ventas que simularemos serán nuestras, por ejemplo, si en un día obtenemos que se venderán 10 propiedades, para cada venta tendríamos cierto porcentaje de probabilidad de que la propiedad esté bajo nuestra administración.
>
>Otro supuesto relevante es que, en este caso, no vamos a simular o contemplar los **costos** de administrar la compañía inmobiliaria, únicamente vamos a evaluar cuáles serían las ganancias potenciales en el mercado inmobiliario de Nueva York, no evaluaríamos por así decirlo la utilidad. 
>
>Vamos a tener dos hipótesis. La primera tiene que ver con que, al momento de analizar los datos, vamos a hacerlo **tomando una muestra de los datos originales**, no tomaremos los datos completos debido al tiempo que tarda en correr. Es por esto que haremos una prueba de hipótesis, en la cual vamos a definir como hipótesis nula que la media de los datos simulados (hechos a partir de la muestra) tienen una media igual a la media de los datos originales/completos. En caso de aceptar la hipótesis nula, podemos decir que los datos simulados son similares a los originales y los podemos tomar como la simulación, en caso de rechazar la hipótesis nula, vamos a volver a generar los datos simulados hasta encontrar unos en los que se acepte la hipótesis nula. 
>
>La segunda hipótesis que queremos probar es que, ya teniendo todas nuestras variables simuladas y determinando las ventas futuras para nuestra inmobiliaria, quisiéramos probar que estas **ganancias tienen una media 
igual a las ganancias simuladas de otros estados (arbitrario).** En caso de que la media sea igual, o que se acepte la hipótesis nula, suponemos que nos será indiferente abrir la inmobiliaria ya sea en Nueva York o en el otro Estado, en el caso contrario, no nos sería indiferente abrir la inmobiliaria en cualquiera de ambos Estados, por lo que analizaríamos que media es mayor y elegiríamos el Estado el cuál dicha media de ventas futuras sea mayor.

```python
def error_distribuciones(a:'Rango inferior',
                         b:'Rango superior',
                        distribuciones:'Distribuciones a probar',
                        datos:'Datos',
                        x:'x del histograma',
                        y:'y del histograma'):
    '''
    Esta función hace algo similar lo que hace fitter, calcula el error de las distribuciones dadas (pueden
    ser todas o un rango), si alguna de ellas arroja error o tarda más de 5 segundos se la salta o no la toma.
    '''
    
    warnings.filterwarnings("ignore")
    
    from interruptingcow import timeout 
    
    error = pd.DataFrame(index = distribuciones, columns = ['Error'])

    for d in distribuciones[a:b]:
        try:
            with timeout(5):
                params = getattr(st, d).fit(datos) 
                y_hat = getattr(st, d)(*params).pdf(x)
                error.loc[d] = ((y - y_hat) ** 2).mean() 
        except (ValueError, RuntimeError):
            continue
    
    return error.sort_values(by = 'Error').head(10)
```

```python
def chi_cuadrada(best:'Nombre distribución',
                datos:'Datos',
                x:'x del histograma',
                y:'y del histograma',
                J1:'Bins del histograma',
                conf:'Confianza'):
    '''
    Esta función realiza la prueba de chi cuadrada de una distribución dada
    con respecto a ciertos datos y con cierta confianza.
    '''
        
    param = getattr(st, best).fit(datos)
    pi = getattr(st, best).pdf(x, *param)
    
    Ei = x*pi

    x2 = np.sum(list(map(lambda Ei, obs_i:(obs_i - Ei)**2 / Ei, Ei, y)))
    print('Valor de chi cuadrado teorico  = ', x2)

    X2 = st.chisquare(y, Ei)
    print('Valor de chi cuadrado librería = ', X2)

    p = len(param) 
    m = J1-p-1 
    
    Chi_est = st.chi2.ppf(q=conf, df=m)
    print('Estadístico de chi_cuadrado = ', Chi_est)
    
    print('La distribución analizada fue: ', best)
```

```python
def kolm_smir(datos:'Datos',
             name:'Nombre distribución'):
    '''
    Función que nos regresa los resultados de una prueba de Kolmogorov-Smirnof
    dada una distribución y unos datos.
    '''
    
    param = getattr(st, name).fit(datos)
    return st.kstest(datos, name, args = (param))
```

```python
def kde_statsmodels_m(x, x_grid, **kwargs):
    """Multivariate Kernel Density Estimation with Statsmodels"""
    kde = KDEMultivariate(x, 
                          bw='cv_ml',  
                          var_type='c', **kwargs)
    return lambda x: kde.pdf([x]) 
```

```python
def get_bandwidth(x:'Datos', 
                 a:'Rango Inferior',
                 b:'Rango Superior',
                 div:'División del rango'):
    grid = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(a, b, div)},
                    cv=50) 

    grid.fit(x[:, None])
    return grid.best_params_
```

```python
def kde_statsmodels_u(x, x_grid, bandwidth, **kwargs):
    """Univariate Kernel Density Estimation with Statsmodels"""
    kde = KDEUnivariate(x)
    kde.fit(bw=bandwidth, **kwargs)
    return lambda x: kde.evaluate(x)
```

```python
def kde_hist_f_u(datos:'Datos',
                bandwidth:'Bandwidth',
                J1:'Bins histograma',
                x:'Linspace'):
    '''
    Esta función nos regresa la función f obtenida por medio del KDE de Statsmodels multivariable 
    para ciertos datos. Además, permite ver el histograma de los datos junto con la función f encontrada
    en una misma gráfica.
    '''
    
    xx = x
    f = kde_statsmodels_u(datos, x, bandwidth)
    
    plt.hist(datos, bins = J1, density = True)
    plt.plot(xx, f(xx), label = 'Función f')
    plt.legend(loc = 'best')
    plt.show()
    
    return f
```

```python
def kde_hist_f(datos:'Datos',
              J1:'Bins histograma',
              x:'Linspace'):
    '''
    Esta función nos regresa la función f obtenida por medio del KDE de Statsmodels multivariable 
    para ciertos datos. Además, permite ver el histograma de los datos junto con la función f encontrada
    en una misma gráfica.
    '''
    
    xx = x
    f = kde_statsmodels_m(datos, xx)
    
    plt.hist(datos, bins = J1, density = True)
    plt.plot(xx, f(xx), label = 'Función f')
    plt.legend(loc = 'best')
    plt.show()
    
    return f
```

```python
def estra_igualmente_espaciado(B:'Número de estratos'):
    """
    Esta función genera variables aleatorias estratificadas de igual manera (con el mismo espacio).
    """
    U2 = np.random.rand(B)
    i = np.arange(0, B)
    v = (U2 + i) / B # Estratificar
    return v
```

```python
def Acep_rechazo_reducc_var(N:'número de aleatorias a generar (aproximado)',
                 a:'rango inferior',
                 b:'rango superior',
                 f:'función objetivo a generar',
                 t:'función que mayora a f',
                 metodo:'función del método de reducción que se usará para generar variables aleatorias que evaluará el método de aceptación rechazo',
                 vav:'número de variables aleatorias a las que se le aplicará el método elegido'):

    '''
    Esta función nos regresa N variables aleatorias de la distribución f dada usando el método de 
    aceptación y rechazo. Además, se aplica el método reducción de varianza elegido en los parámetros
    al momento de generar las variables aleatorias R2. Se generarán variables aleatorias con reducción 
    de varianza las veces que sea necesario hasta llegar a los N valores especificados. 
    '''
    puntos_aceptados = 0
    j = 0
    x = np.zeros(N)
    while puntos_aceptados < N:
        if j == 0:
            R2_a = metodo(vav) # Si j es 0, genera nuevas aleatorias con el método elegido.
        R2 = R2_a[j]           # Esta aleatorias con reducción de varianza se evalúan con aceptación rechazo.
        R1 = np.random.uniform(a, b)
        f_x = f(R1) 
        t_x = t(R1)
        condition = R2 * t_x <= f_x
        [plt.plot(R1,R2*t_x,'ob') if condition else plt.plot(R1,R2*t_x,'o')]           
        if condition:
            x[puntos_aceptados] = R1
            puntos_aceptados += 1
        else:
            None
        j += 1
        if j == vav - 1: # Si aún no se llega a la N deseada, j se hace 0 y se vuelven a generar aleatorias con reducción de varianza.
            j = 0        # Es decir, puede que con 1 set de v.a. con reducción se llegue a N o no, si no se llega, se genera otro set de v.a. con reducción.
    plt.show()
    
    return x
```

```python
def Acep_rechazo_Bestrr(N:'número de aleatorias a generar (aproximado)',
                 a:'rango inferior',
                 b:'rango superior',
                 f:'función objetivo a generar',
                 t:'función que mayora a f',
                 r2_no_est:'variables aleatorias a estratificar'):

    '''
    Esta función nos regresa N (aproximadamente) puntos o variables aleatorias de la distribución f
    dada usando el método de aceptación y rechazo visto en clase. Se aplica el método de estratos
    igualmente estratificados al generar las variables aleatorias R2.
    '''
    
    R2_a = estra_igualmente_espaciado(r2_no_est)
    puntos_aceptados = 0
    j = 0
    x = np.zeros(N)
    #print(R2_a)
    while puntos_aceptados < N:
        R2 = R2_a[j]
        #print(R2)
        R1 = np.random.uniform(a, b)
        f_x = f(R1) 
        t_x = t(R1)
        condition = R2 * t_x <= f_x
        [plt.plot(R1,R2*t_x,'ob') if condition else plt.plot(R1,R2*t_x,'o')]           
        if condition:
            x[puntos_aceptados] = R1
            puntos_aceptados += 1
        else:
            None
        j += 1
        if j == r2_no_est - 1:
            break
    plt.show()
    
    return x
```

```python
def generar_aleatorias_reducc_var(va:'Aleatorias a generar',
                               f:'Función de distribución',
                               v:'Valor máximo (aproximado) de la distribución',
                               xx:'Valores de x a evaluar (linspace)',
                               datos:'Datos sobre los cuales se generarán las aleatorias',
                               metodo:'función que genera aleatorias con el método de reducción de varianza deseado',
                               varv:'# de variables aleatorias a las que se le aplicará el método de reducción de varianza'):
    '''
    Esta función nos genera variables aleatorias con varianza reducida, especificando
    el método de reducción de varianza a aplicar en los argumentos de la función, de una 
    distribución f dada aplicando el método de aceptación y rechazo. Se usa una constante
    como función que mayora a f.
    '''
    
    max_f = f(optimize.fmin(lambda x: -f(x), v, disp=False))
    
    t = lambda x: max_f  # Graficar la constante
    t_2 = lambda x: max_f * np.ones([len(x)])
    
    # Gráficas de las funciones
    plt.plot(xx,f(xx),label='t(x)') # Función f
    plt.plot(xx,t_2(xx),label='t(x)') # Constante
    plt.legend(loc = 'best')
    
    aleat = Acep_rechazo_reducc_var(va, np.amin(datos), np.amax(datos), f, t, metodo, varv)
    
    return aleat
```

```python
def generar_aleatorias(f:'Función de distribución',
                      v:'Valor máximo gráfica',
                      xx:'Linspace',
                      datos:'Datos',
                      va:'Aleatorias a generar'):
    '''
    Esta función nos genera variables aleatorias con varianza reducida de una 
    distribución f dada, aplicando el método de aceptación y rechazo.
    '''
    
    max_f = f(optimize.fmin(lambda x: -f(x), v, disp=False))
    
    t = lambda x: max_f  # Graficar la constante
    t_2 = lambda x: max_f * np.ones([len(x)])
    
    # Gráficas de las funciones
    plt.plot(xx,f(xx),label='t(x)') # Función f
    plt.plot(xx,t_2(xx),label='t(x)') # Constante
    plt.legend(loc = 'best')
    
    v_a_price = Acep_rechazo_Bestrr(va, 0, np.amax(datos), f, t, 8000)
    v_a_price = v_a_price[v_a_price != 0]
    
    return v_a_price
```

```python
def int_conf(confianza:'Confianza',
            datos:'Datos a sacar el intervalo'):
    '''
    Esta función nos regresa el intervalo de confianza y la media real de ciertos datos.
    '''
    
    i_price = st.t.interval(confianza, len(datos)-1, loc=np.mean(datos), scale=st.sem(datos))

    return i_price, np.mean(datos)
```

```python
def prueba_hip(meand:'Media real de los datos',
              datos:'Datos a probar',
              confianza:'Significancia (0.05)'):
    '''
    Esta función no realiza la prueba de hipótesis de si nuestros datos tienen
    una media igual al valor que se especifique. 
    '''
    
    # Prueba de hipótesis de muestra
    ho = st.ttest_1samp(datos, meand) # Prueba de hipótesis de una muestra 
    print('La prueba de hipótesis arroja como resultado\n',ho)
    
    # Calculamos el estadístico normalizado
    t = (np.mean(datos) - meand)/(np.std(datos) / np.sqrt(len(datos)))
    print('Cálculo del estadístico de prueba teórico=', t)
    
    # Región de rechazo
    cuantil = st.t(len(datos)-1).cdf(1-confianza/2) 
    print(f'Región de rechazo = (t<{-cuantil}) U (t>{cuantil})')
```
